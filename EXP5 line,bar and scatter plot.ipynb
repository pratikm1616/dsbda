```python
## plot2
import matplotlib.pyplot as plt
names = ['a', 'b', 'c']
values = [1, 10, 100]
plt.subplot(141)
plt.bar(names, values)
plt.subplot(142)
plt.scatter(names, values)
plt.subplot(143)
plt.plot(names, values, linewidth=5.0)
plt.subplot(144)
plt.plot(names, values, '--')
plt.show()
```

```python
## Plot3
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
pstore = pd.read_csv("C:/Users/Mangesh/Downloads/Data sets-20240124T042619Z-001/Data sets/dataset_Facebook.csv")
pstore.head(10)
#Create a distribution plot for rating
sns.displot(pstore.like)
#Adding dark background to the graph
plt.style.use("dark_background")
# Change the number of bins
sns.displot(pstore.like, bins=20, kde = False, color='red')
```

```python
sns.histplot(pstore.like, bins=20, kde = False, color='red')
#Plotting the scatter plot
plt.scatter(pstore.share, pstore.like)
# Plotting the same thing now using a jointplot
sns.pairplot(pstore[['like', 'share', 'comment','Total Interactions']])
plt.show()
```
